## DataFrame, df2

```python
data2 = [
    {"Name":"Zizipho",
    "Phone":843774,
    "Course":"Actuarial Science",
    "Current Role":"Student"},
    
    {"Name":"Saumu",
    "Phone":19843774,
    "Course":"Maths",
    "Current Role":"Tutor"},
    
    {"Name":"Kiprono",
    "Phone":454774,
    "Course":"Data Science",
    "Current Role":"Tutor"}
]

df2=pd.DataFrame(data2)
df2
```

```python
#outer merge
pd.merge(df1, df2, how='outer',on="Name")
```

```python
#inner
pd.merge(df1, df2, how='inner', on="Name")#,left_index=True, right_index=True)
```

```python
#left
pd.merge(df1, df2, how='left', on="Name")
```

```python
#right
pd.merge(df1, df2, how='right',on = "Name")
```
